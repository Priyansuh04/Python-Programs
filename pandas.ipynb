### Pandas-DataFrame And Series 
#### Pandas is a powerful data manipulation library in python, widely used for  data analysis and data cleaning. it provides two primary data structures: series and dataframe. a series is a one dimensional array-like object, while a dataframe is two-dimensional, size-mutable,and potenially heterogenous tabular data structure with labeled axes(rows and columns). 

```python
import pandas as pd
```

```python
## Series 
### A series is a one dimensional array-like object that can hold any data type. it is similar to a column in a table.
import pandas as pd
data=[1,2,3,4,5]
s=pd.Series(data)
print("Series\n",s)
```

```python
## Create a Series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)

```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
## create a dataframe from a dictionary of list

data={
    "name":['krish','john','jack'],
    "age" :[24,30,36],
    "city":['bangalore','mumbai','pune']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a dataframe from a list of dictionaries

data=[
    {"name":'krish',"age":24,"city":'bangalore'},
    {"name":'john', "age":30,"city":'mumbai'},
    {"name":'ishan',"age":35,"city":'pune'},
    {"name":'jack', "age":40,"city":'delhi'}
]

daf=pd.DataFrame(data)
print(daf)
print(type(daf))
```
