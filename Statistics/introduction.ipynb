## Types of statistics 


### 1 Descriptive Statistics 
Descriptive Statistics involves methods for summarizing and organizing data to make it understandable. this type of statistics helps to describe the basic feature of the data in a study. 

### 2 Inferential Statistics 
Inferential satistics involves methods for making predictions or inferences about a population based on a sample of data. it allows for hypothesis testing, estimation, and drawing conclusions.

## Types of sampling Techniques
### 1 Probality Sampling , 2 Non Probality Sampling

### Probality Sampling 

### a. Simple random sampling 
Every member of the population has equal chance of being selected.

### b. Systematic sampling
Select every nth member of the population after a random starting point.

### c. Statisfied sampling 
Divide the population into strata (groups) based omn specific charteristic and then randomly sampling from each strata.

### d. Cluster sampling
Divide the population into clusters, randomly selecting clusters,then sampling all the members from the selected clusters.

### e. Milti stage sampling
Combining several sampling methods usually involves selecting clusters,then randomly sampling with in those cluster.
