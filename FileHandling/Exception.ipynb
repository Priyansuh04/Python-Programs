## Understanding Exception
Exception handling in python allows you to handles errors gracefully and take coorective actions without stopping the execution of the program. This lesson will cover the basics of exceptions, including how to use try, except, else, and finally blocks.


## What are Exception?
## Exception are events that disrupt the normal flow of a program. they occur when an error is encountered during program execution. coomon exceptions includes:
#### ZeroDivisionError: Dividing by zero.
#### FileNotFoundError: File not found
#### ValueError: invalid value.
#### TypeError: Invalid type.
#### NameError: Not defined

```python
try:
    num=int(input("enter the number"))
    result=10/num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("enter denominator greater than 0")  
```

```python
## Try, except, else block
try:
    num =int(input("enter the number"))
    result=10/num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("you can't divide by zero")  
except Exception as ex1:
    print(ex1)
else:
    print(f"the result is {result}")          
```

```python
## Try, except, else and finally

try:
    num =int(input("enter the number"))
    result=10/num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("you can't divide by zero")  
except Exception as ex1:
    print(ex1)
else:
    print(f"the result is {result}")   
finally:
    print("Execution complete.")    
```
