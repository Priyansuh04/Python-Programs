## File operation- Read and Write Files
file handling is a cruical part of any programming language. python provides bulit-in functions and methods to read and write to files, both text and binary.
this lesson will cover the basics of handling, including reading and writing text files and binary files.



```python
## read a whole file

with open("sample.txt","r") as file:
    content=file.read()
    print(content)
```

```python
with open("sample.txt",'r') as file:
    for line in file:
        print(line.strip()) ## sstrip() removes the newline character
```

```python
## writing a file(overwriting)

with open("sample.txt","w") as file:
    file.write("hello wprld\n")
    file.write("this is a new line.")
```

```python
## write a file(without overwriting)

with open("sample.txt","a") as file:
    file.write("append operation taking place\n")    

```

```python
## Writing a list of lines to a file 
lines=['first line\n','second line\n','third line\n']
with open("sample.txt","a") as file:
    file.writelines(lines)
```

```python
## Binary files 
 
## writing to a binary file 
data =b"x00\x01\x02\x03\x04"
with open("example.bin","wb") as file:
    file.write(data)
```

```python
## Reading a binary file
with open("example.bin","rb") as file:
    content=file.read()
    print(content)
```

```python
### read the content from a source text file and write to a destination text file
# copying a text file
with open("sample.txt","r") as source_file:
    content=source_file.read()
    
with open("sample.txt","w") as destination_file:
    destination_file.write(content)
```

```python
## Read a text file and count the number of lines, words, and characters.
# counting lines, words and characters in a text file
def count_text_file(file_path):
    with open(file_path,"r") as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines)
        char_count=sum(len(line) for line in lines)
    return line_count, word_count, char_count 

file_path="sample.txt"
lines, words, charcters =count_text_file(file_path)   
print(f"lines:{lines}, words:{words}, characters:{charcters}")
```

```python
#Checking if a path is a file or directory 
import os 

path="sample.txt"
if os.path.isfile(path):
    print(f"the path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"the path '{path}' is a directory.")
else:
    print(f"the path '{path}' is a neither a file nor a directory.")    
```
