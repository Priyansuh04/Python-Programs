```python
## map() ## applies a function to all items in a list 
numbers =[1,2,3,4,5,6,]
def square(numbers):
    return numbers**2

square(4)
```

```python

# def square(x):
#     return x * x

nums = [1, 2, 3, 4]

print(list(map(square,nums)))
```

```python
## Lmabda function with map
numbers =[1,2,3,4,5,6,]
list(map(lambda x:x**2,numbers))

```

```python
## Map multiple iterable 

num1=[1,2,3,4]
num2=[5,6,7,8]

result =list(map(lambda x,y:x+y,num1,num2))
print(result)
```

```python
## map() to convert a list of string to integers
## use map to convert strings to integers

str_numbers =['1','2','3','4','5']
int_numbers =list(map(int,str_numbers))

print(int_numbers)
```

```python
words =['apple','banana','cherry']
upper_words =list(map(str.upper,words))
print(upper_words)
```

```python
def get_name(person):
    return person["name"]
    

people =[
    {"name":"krish","age":24},
    {"name":"jack","age":32}
]

list(map(get_name,people))

```
