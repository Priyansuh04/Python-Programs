**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Using For Loop
factorial= int(input("enter the number :"))
fact = 1
for i in range(1,factorial+1):
    fact=fact*i

print("factorial number is :",fact)    

```

```python
# Using While Loop
factorial= int(input("enter the number :"))
fact = 1
i=1
while i<=factorial:
    fact = fact*i
    i=i+1
 
print("factorial number is :",fact)
```
