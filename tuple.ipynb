### Introduction to tuples 
### Explanation:
### Tuples are ordered collection of item that are immutable. they are similar to lists, but their immutability makes them different. 

```python
## creating a tuple 
empty_tuple =()
print(empty_tuple)
print(type(empty_tuple))
```

```python
## immutable nature of tuples
## tuple are immutable, meaning their elements cannot be changed.

lst =[1,2,3,4,5]
print(lst)

lst[1]="krish"
print(lst)
```

```python
numbers = ([1,2,3,4,5,6]) 
numbers[1] ="krish"
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers[1]="krish"
```

```python
## tuple methods
numbers = tuple([1,2,3,4,5,6])
print(numbers.count(1))
print(numbers.index(3))
```

```python
## packing and unpacking tuple
##packing
packed_tuple =1,"hello",3.
print(packed_tuple)
```

```python
## unpacking a tuple
a,b,c =packed_tuple

print(a)
print(b)
print(c)
```

```python
## unpacking with * 
numbers = (1,2,3,4,5,6)
first,*middle,last =numbers
print(first)
print(middle)
print(last)
```

```python
## nested tuple
## nested list
lst = [[1,2,3,4],[6,7,8,9],[1,"hello",3.14,"c"]]
print(lst[0])
print(lst[0][2])
```

```python
print(lst[0][0:3])
```

```python
lst = [[1,2,3,4],[6,7,8,9],(1,"hello",3.14,"c")]
print(lst[2][0:3])
```

```python
## access the element inside a tuple 
nested_tuple =((1,2,3),("a","b","c"),(True,False))
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end =" ")
    print()    
```

conclusion

Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of item is required. they are commonly used in data structure ,function arguments and return values, and as dictionary keys. understanding how to leverage tuples effectively can improve the efficiency and readability of your python code.
