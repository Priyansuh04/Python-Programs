### Introduction to tuples 
### Explanation:
### Tuples are ordered collection of item that are immutable. they are similar to lists, but their immutability makes them different. 

```python
## creating a tuple 
empty_tuple =()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst =list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers 
```

```python
list((1,2,3,4,5))
```
