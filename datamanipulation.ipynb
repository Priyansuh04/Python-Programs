```python
## Renaming columns
df=df.rename(columns={"Age":"Year"})
df.head()
```

```python
## Change datatypes
df["Year_new"]=df["Year"].astype(float)
df.head()
```

```python
df["Year_new"]=df["Year"].apply(lambda x:x*2)
```

```python
## Data Aggregating And Grouping
df.head()
```

```python
grouped_mean=df.groupby("Occupation")["Sleep Duration"].mean()
```

```python
print(grouped_mean)
```

```python
grouped_sum=df.groupby(["Occupation","Gender"])["Year"].sum()
print(grouped_sum)
```

```python
grouped_mean=df.groupby(["Occupation","Gender"])["Year"].mean()
print(grouped_mean)
```

```python
## Aggregate multiple functions
grouped_agg=df.groupby("Gender")["Year"].agg(["mean","sum","count"])
```

```python
grouped_agg
```

```python
## Merging and joining dataframes
## Create sample dataframes
df1=pd.DataFrame({"key":["a","b","c"],"value1":[1,2,3]})
df2=pd.DataFrame({"key":["a","b","d"],"value2":[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge dataframe on the 'key column'
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
