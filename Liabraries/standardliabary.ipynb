## Standard library overview 

python's standard library is a vast coolection of modules and packages that come bundled with python, providing a wide range of functionalities out of the box. here's an overview of some of the most commonly used modules and packages in the python standard library.


```python
import array
arr=array.array('i',[1,2,3,4,5])
print(arr)
```

```python
import math
(math.sqrt(16))
print(math.pi)
```

```python
## Random
import random 
print(random.randint(1,10))
print(random.choice(["apple","banana","cherry"]))
```

```python
## File and directory
import os
print(os.getcwd())
```

```python
## high level operation on files and collection 
import shutil
# shutil.copyfile('source.txt','destination.txt')
```

```python
## Data serialization 
import json
data ={'name':'krish','age':23}

json_str=json.dumps(data)
print(json_str)
print(type(json_str))
```

```python
parsed_data=json.loads(json_str)
print(type(parsed_data))
```

```python
## csv
import csv 
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['krish',24])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)  
    for row in reader:
        print(row)
```

```python
## datetime

from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
## time
import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
## regular expression
import re
pattern=r'\d+'
text='there are 123 apples 456'
match=re.search(pattern,text)
print(match.group())
```
