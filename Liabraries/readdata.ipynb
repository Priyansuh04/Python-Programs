## Reading data from different sources

```python
import pandas as pd
from io import StringIO
Data='''
{
  "employee_name": "james",
  "email": "james@gmail.com",
  "job_profile": [
    {"title1": "team lead", "title2": "Sr. developer"}
  ]
}'''
pd.read_json(StringIO(Data))
```

```python
df=pd.read_json(StringIO(Data))
print(df)
```

```python
df.to_json
```

```python
df.to_json(orient="index") 
```

```python
df.to_json(orient="records")
```

```python
import pandas as pd

data = [
    [1, 10, 100],
    [2, 20, 200],
    [3, 30, 300],
    [4, 40, 400]
]

# create dataframe
df = pd.DataFrame(data, columns=["A", "B", "C"])

# write to csv
df.to_csv("numbers.csv", index=False)

# read from csv
daf = pd.read_csv("numbers.csv")
print(daf)

```

```python
daf.head()
```

```python
df = pd.read_csv("numbers.csv")
print(df)
```

```python
import pandas as pd

# read csv
df = pd.read_csv("numbers.csv")

# save as excel
df.to_excel("numbers.xlsx", index=False)

# read excel
df_excel = pd.read_excel("numbers.xlsx")
print(df_excel)

```
