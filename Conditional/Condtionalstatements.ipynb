```python
## if statement
## if statement execute if the block of code when condition is true

age = 20

if age >= 18:
    print(" you are adult ")

```

```python
## else statement
## the else statement is execute a block of code if the condition in the if satatement is false

age = 16
if age >= 18:
    print(" you are adult ")
else:
    print(" you are child ")    
```

```python
## elif satement            
## the elif satement is allows to check multiple conditions. it stands for "else if"

age =10
if age >= 18:
    print(" you are adult ")
elif age >= 10:
    print("you are teenager")
else:
    print(" you are child ") 


```

```python
## nested conditional statement 
## you can one or more if , elif or else statement inside another if , elif , or else statement to create nested conditional statement.

## number odd , even , negative
num =int(input("enter the number "))

if num >0:
    print("the number is positive")
    if num%2==0:
        print("the number is even ")
    else:
        print("the number is odd ")
        
else:
    print("the number is negative ")            
```

```python
## pratical example 
## determine if a year is a leap year using nested condition statement 

year = int(input("enter the year "))

if year % 4 == 0 :
    if year % 100 == 0:
        if year % 400 == 0:
            print(" this  leap year ",year)
        else:
            print("this year is not leap year",year)    
    else:
        print(" this  leap year ",year)
else:
     print("this year is not leap year",year) 
             
```
