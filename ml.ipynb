## PRACTICAL 4: pandas Data Handling

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Marks": [78, 85, 90]
}

df = pd.DataFrame(data)
print(df)

# Data exploration
print(df.head())
print(df.info())
print(df.describe())
```

## PRACTICAL 6: Data Slicing & Indexing

```python
print(df["Marks"])
print(df.loc[0])
print(df.iloc[1])
```

## PRACTICAL 7: Data Visualization (matplotlib)

```python
import matplotlib.pyplot as plt

plt.plot(df["Name"], df["Marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.show()
```

## PRACTICAL 8: Data Visualization (seaborn)

```python
import seaborn as sns

sns.barplot(x="Name", y="Marks", data=df)
plt.title("Marks Bar Chart")
plt.show()
```

## PRACTICAL 9: Simple Regression Model

```python
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, y)
pred = model.predict([[6]])
print("Prediction:", pred)
```
