```python
## create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set =set([1,2,3,4,4,4,5,6,6,6])
print((my_empty_set))
```

```python
## basics ses operation
## adding and removing elements 
my_set.add(6)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
## remove the elements from a set 
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## pop method 
my_set = {1,2,3,4,5}
remove_element =my_set.pop()
print(remove_element)
print(my_set)
```

```python
## clear all the elements
my_set = {1,2,3,4,5}
my_set.clear()
print(my_set)
```

```python
## set membership test 
my_set = {1,2,3,4,5}
print( 3 in my_set)
print( 10 in my_set)
```

```python
## mathematical operation 
set1={1,2,3,4,5}
set2={4,5,6,7,8}

## union
union_set =set1.union(set2)
print(union_set)
```

```python
## intersection 
intersection_set=set1.intersection(set2)
print(intersection_set)

```

```python
set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5}
set2={4,5,6,7,8}

## difference
print(set1.difference(set2))


```

```python
##set methods
set1 ={1,2,3,4,5}
set2 ={3,4,5}

## is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```
