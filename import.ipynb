## Import modules in python: Modules and Packages 
In python, modules and packages help organize and reuse code. Here's a comprehension guide on how to import them.

```python
import math 
math.sqrt(16)
```

```python
from math import sqrt,pi
print(math.sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math import *
print(sqrt(16))
print(pi)
```

```python
import os

# Create the 'package' directory if it doesn't exist
if not os.path.exists('package'):
    os.makedirs('package')

# Create an empty __init__.py file inside 'package' to make it a package
with open('package/__init__.py', 'w') as f:
    pass

# Create the 'math.py' file inside 'package' with the addition function
math_code = """def subtraction(a, b):
    return a - b
"""
with open('package/math.py', 'w') as f:
    f.write(math_code)

print("Local 'package' with 'math.py' created successfully.")
```

```python
import os

# Create the 'package' directory if it doesn't exist
if not os.path.exists('package'):
    os.makedirs('package')

# Create __init__.py to make it a package
with open('package/__init__.py', 'w') as f:
    pass

# Create math.py with basic math functions
math_code = """
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Division by zero not allowed"
    return a / b
"""

with open('package/math.py', 'w') as f:
    f.write(math_code)

print("Local 'package' with 'math.py' created successfully.")
```

```python
from package.math import addition
addition(2,3)
```

```python
from package import math
math.addition(4,5)
```

```python
from package.math import subtract
subtract(2,3)
```

```python
from package import math
math.subtract(4,5)
```
